```python
import pandas as pd
from bs4 import BeautifulSoup

all_links = []

for i in range(1,314):
    file_names = [str(i)+'.html'] 
    for file_name in file_names:
        with open("/Users/macbookpro/Desktop/lxc_2024/lxc_2023/git/NFT_Crypto/scripted website/"+file_name, 'r', encoding='utf-8') as file:
            html_content = file.read()

        soup = BeautifulSoup(html_content, "html.parser")
        links = [a['href'] for a in soup.find_all('a', class_='dapp-name-link-comp')]
        all_links.extend(links) 
    


print(all_links)
print(len(all_links))


```

```python
from collections import defaultdict
directory_path = "/Users/macbookpro/Desktop/lxc_2024/lxc_2023/git/NFT_Crypto/scripted website/"
# Initialize a defaultdict to track the files each link appears in
link_file_map = defaultdict(list)

for i in range(1, 314):
    file_name = f"{i}.html"
    try:
        with open(f"/Users/macbookpro/Desktop/lxc_2024/lxc_2023/git/NFT_Crypto/scripted website/{file_name}", 'r', encoding='utf-8') as file:
            html_content = file.read()

        soup = BeautifulSoup(html_content, "html.parser")
        links = [a['href'] for a in soup.find_all('a', href=True)]
        
        for link in links:
            link_file_map[link].append(file_name)

    except FileNotFoundError:
        print(f"File {file_name} not found.")

# Now identify duplicates across all files
duplicates = {link: files for link, files in link_file_map.items() if len(files) > 1}

print(f"Total duplicates across all files: {len(duplicates)}")

# Optionally, print duplicates and their associated files for verification
for link, files in duplicates.items():
    print(f"Link {link} appears in files: {', '.join(files)}")
```
