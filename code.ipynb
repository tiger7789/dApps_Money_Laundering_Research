```python
import pandas as pd
from bs4 import BeautifulSoup

all_links = []

for i in range(1,314):
    file_names = [str(i)+'.html'] 
    for file_name in file_names:
        with open("/Users/macbookpro/Desktop/research/"+file_name, 'r', encoding='utf-8') as file:
            html_content = file.read()

        soup = BeautifulSoup(html_content, "html.parser")
        links = [a['href'] for a in soup.find_all('a', class_='dapp-name-link-comp')]
        all_links.extend(links) 


print(all_links)
print(len(all_links))

```

```python
#convert to csv

df = pd.DataFrame(all_links, columns=['Links'])
csv_file_path = '/Users/macbookpro/Desktop/research/extracted_links.csv'
df.to_csv(csv_file_path, index=False)

csv_file_path
```

```python
import time
import webbrowser
import pyautogui as p

def open_close(url="https://dappradar.com/dapp/kai-ching"):
    #open url
    webbrowser.open(url)
    #wait for 10 seconds
    time.sleep(15)  # Adjust this based on how quickly your browser loads the page
    
    #save the page
    p.hotkey("command", "s")
    time.sleep(5)
    p.press("enter")
    time.sleep(3)
    p.press("enter")
    time.sleep(3)

    #close tab 
    p.keyDown("command")
    p.press("w")
    p.keyUp("command")
    print("tab closed")

open_close()

```
